```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
final_df = pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
final_df.columns
```

```python
final_df.Country.value_counts()
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_val = final_df.Country.value_counts().values
```

```python
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

# Observation: 
Zomato maximum records or transactions are from India, after that USA and then United Kingdom. 

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

# Observation:
1. Rating in between 4.5 to 4.9 ---> Excellent
2. Rating in between 4.0 to 4.4 ---> Very good
3. Rating in between 3.5 to 3.9 ---> Good
4. Rating in between 3.0 to 3.4 ---> Average
5. Rating in between 2.5 to 2.9 ---> Average
6. Rating in between 2.0 to 2.4 ---> Poor
7. Most of the people don't give rating to this app

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating", y="Rating count", data=ratings)
```

```python
sns.barplot(x="Aggregate rating", y="Rating count", hue='Rating color', data=ratings, palette=['Blue','Red','Orange','Yellow','Green','Green'])
```

# Observation:
1. Not rated counts are the most
2. Maximum number of ratings are between 3.0 to 3.9

```python
sns.countplot(x='Rating color',data=ratings,hue='Rating color',palette=['Blue','Red','Orange','Yellow','Green','Green'])
```
